###Q1.What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample to infer that a certain condition is true for an entire population.

##Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H₀) is a statement that there is no effect or no difference. It represents the status quo.

The alternative hypothesis (H₁ or Hₐ) is a statement that there is an effect or a difference. It is what the researcher aims to prove.

###Q3What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It is usually set at 0.05 (5%) and determines the threshold for statistical significance.

##Q4.What does a P-value represent in hypothesis testing?
The P-value is the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true.

###Q5.How do you interpret the P-value in hypothesis testing?
If P ≤ α, reject the null hypothesis (H₀) → evidence suggests an effect.

If P > α, fail to reject the null hypothesis → insufficient evidence to conclude an effect.

###Q6.What are Type 1 and Type 2 errors in hypothesis testing?

Type 1 Error (False Positive): Rejecting H₀ when it is true (α risk).

Type 2 Error (False Negative): Failing to reject H₀ when Hₐ is true (β risk).

###Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

###One-tailed test:
Tests for an effect in one direction (e.g., greater than or less than).

###Two-tailed test:
Tests for an effect in both directions (e.g., not equal to).

###Q8.What is the Z-test, and when is it used in hypothesis testing?
A Z-test is used when testing hypotheses about population means when the sample size is large (n ≥ 30) and the population variance is known.

###Q10.What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is a probability distribution used when the sample size is small (n < 30) and the population standard deviation is unknown.

###Q11.What is the difference between a Z-test and a T-test?

Z-test: Used when population variance is known and n ≥ 30.

T-test: Used when population variance is unknown and n < 30.

###Q12.What is the T-test, and how is it used in hypothesis testing?
A T-test is used to compare sample means when the population standard deviation is unknown. It helps determine if two samples come from populations with the same mean.

###Q13.What is the relationship between Z-test and T-test in hypothesis testing?
Both tests compare sample means to a population mean, but a T-test is used when the standard deviation is unknown, whereas a Z-test is used when it is known.

###Q14.What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval (CI) is a range of values within which the true population parameter is expected to lie with a given probability (e.g., 95% CI).

###Q15.What is the margin of error, and how does it affect the confidence interval?
The margin of error is the maximum expected difference between the sample estimate and the true population value. A larger margin results in a wider confidence interval.

###Q16.How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem is used to update probabilities based on new evidence. It is significant in fields like machine learning and medical diagnostics.

###Q17.What is the Chi-square distribution, and when is it used?
The Chi-square distribution is used in hypothesis testing for categorical data, such as goodness-of-fit tests and independence tests.

###Q18.What is the Chi-square goodness-of-fit test, and how is it applied?
The Chi-square goodness-of-fit test determines if observed categorical data matches expected distributions. It compares observed frequencies with expected frequencies.

###Q19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is a probability distribution that arises in variance analysis and is always right-skewed. It is used in hypothesis testing to compare variances between two or more groups.

Uses of the F-distribution in hypothesis testing:
ANOVA (Analysis of Variance): To test if multiple group means are equal.

F-test for equality of variances: To compare the variances of two populations.

Regression analysis: To assess the overall significance of a regression model.

###Q20.What is an ANOVA test, and what are its assumptions?
The ANOVA (Analysis of Variance) test is a statistical method used to compare means of three or more groups to determine if they come from the same population. It is based on the F-distribution.

Assumptions of ANOVA:
Independence: Observations in each group are independent.

Normality: Data in each group should be approximately normally distributed.

Homogeneity of variances (Homoscedasticity): The variance among the groups should be approximately equal.

###Q21.What are the different types of ANOVA tests?
There are several types of ANOVA tests, depending on the number of factors being analyzed:

One-way ANOVA: Compares the means of one independent variable with three or more groups.

Two-way ANOVA: Examines the effect of two independent variables on a dependent variable.

Repeated Measures ANOVA: Used when the same subjects are tested under multiple conditions.

MANOVA (Multivariate ANOVA): Used when there is more than one dependent variable.

###Q22.What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test that compares the variances of two or more groups. It is used in ANOVA and regression analysis to test the equality of variances or the significance of model parameters.

Key uses of the F-test in hypothesis testing:
In ANOVA, the F-test determines if there is a statistically significant difference between group means.

In regression analysis, the F-test evaluates whether a regression model significantly fits the data.



```python
###PRACTICAL
#Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test
    return z_score, p_value

# random data
np.random.seed(42)
sample_data = np.random.normal(50, 10, 30)
pop_mean = 52
pop_std = 10

z, p = one_sample_z_test(sample_data, pop_mean, pop_std)
print(f"Z-score: {z:.2f}, P-value: {p:.4f}")
if p < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

```python
##Q2.Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm

def plot_z_test(z_score, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    critical_value = norm.ppf(1 - alpha/2)  # Two-tailed
    plt.figure(figsize=(8, 5))

    # Pass data as keyword arguments
    sns.lineplot(x=x, y=y, color="blue") #changed this line to pass x and y as keyword arguments

    # Shade rejection regions
    plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color='red', alpha=0.5)

    # Plot Z-score
    plt.axvline(z_score, color='black', linestyle="dashed", label=f"Z-score = {z_score:.2f}")
    plt.axvline(critical_value, color='red', linestyle="dashed", label=f"Critical Z = {critical_value:.2f}")
    plt.axvline(-critical_value, color='red', linestyle="dashed")

    plt.title("Z-Test Decision Region")
    plt.legend()
    plt.show()

plot_z_test(z)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample = [100, 102, 98, 101, 99, 100, 97, 103, 99, 98]
population_mean = 100
population_std = 2
alpha = 0.05
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_critical = norm.ppf(1 - alpha / 2)
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(f"Sample mean: {sample_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
print(f"Critical Z-values: ±{z_critical}")

# Decision
if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Z-distribution')
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z = {z_score:.2f}')

plt.axvline(-z_critical, color='red', linestyle='--', label=f'-Z_critical = {-z_critical:.2f}')
plt.axvline(z_critical, color='red', linestyle='--', label=f'+Z_critical = {z_critical:.2f}')

# Plot formatting
plt.title('Two-Tailed Z-Test Visualization')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ✅ Run this in Google Colab
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu_0, mu_1, sigma, n, alpha=0.05):

    se = sigma / np.sqrt(n)
    z_crit = norm.ppf(1 - alpha/2)
    x_crit_low = mu_0 - z_crit * se
    x_crit_high = mu_0 + z_crit * se

    # x values for plotting
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    y_h0 = norm.pdf(x, mu_0, se)
    y_h1 = norm.pdf(x, mu_1, se)

    # Create the plot
    plt.figure(figsize=(12, 6))
    plt.plot(x, y_h0, label=f'H₀: μ = {mu_0}', color='blue')
    plt.plot(x, y_h1, label=f'H₁: μ = {mu_1}', color='green')

    # Shade Type I error
    plt.fill_between(x, y_h0, where=(x <= x_crit_low) | (x >= x_crit_high),
                     color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II error
    plt.fill_between(x, y_h1, where=(x > x_crit_low) & (x < x_crit_high),
                     color='orange', alpha=0.4, label='Type II Error (β)')

    # Mark critical boundaries
    plt.axvline(x_crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(x_crit_high, color='black', linestyle='--')


    plt.title("Visualization of Type I (α) and Type II (β) Errors in Z-Test")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Calculate and display beta
    beta = norm.cdf(x_crit_high, mu_1, se) - norm.cdf(x_crit_low, mu_1, se)
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the Test (1 - β): {1 - beta:.4f}")

plot_type1_type2_errors(mu_0=100, mu_1=102, sigma=5, n=30, alpha=0.05)
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def independent_t_test():
    print("Independent T-test")
    group1 = [25, 30, 28, 35, 33, 31]
    group2 = [22, 27, 24, 26, 25, 23]

    t_stat, p_value = stats.ttest_ind(group1, group2)

    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.3f}")
    if p_value < 0.05:
        print("Result: Significant difference between the two groups.")
    else:
        print("Result: No significant difference between the two groups.")
independent_t_test()
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def paired_t_test():
    print("Paired Sample T-test")
    before = [88, 75, 90, 85, 80]
    after =  [92, 78, 88, 86, 84]

    t_stat, p_value = stats.ttest_rel(before, after)

    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.3f}")
    if p_value < 0.05:
        print("Result: Significant change after treatment.")
    else:
        print("Result: No significant change after treatment.")

    # Visualization
    x = np.arange(len(before))
    plt.figure(figsize=(8,4))
    plt.plot(x, before, marker='o', label='Before')
    plt.plot(x, after, marker='o', label='After')
    plt.title('Paired Sample Comparison')
    plt.xlabel('Subjects')
    plt.ylabel('Scores')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
paired_t_test()
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python
def compare_z_t_tests():
    print("\n--- Z-test vs T-test ---")
    np.random.seed(1)
    sample = np.random.normal(loc=50, scale=10, size=30)
    population_mean = 52
    population_std = 10  # Known for Z-test

    # Z-test
    z_stat = (np.mean(sample) - population_mean) / (population_std / np.sqrt(len(sample)))
    z_p = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # T-test
    t_stat, t_p = stats.ttest_1samp(sample, population_mean)

    print(f"Z-test: Z = {z_stat:.3f}, p = {z_p:.3f}")
    print(f"T-test: t = {t_stat:.3f}, p = {t_p:.3f}")

    if z_p < 0.05: print("Z-test: Significant difference.")
    else: print("Z-test: No significant difference.")

    if t_p < 0.05: print("T-test: Significant difference.")
    else: print("T-test: No significant difference.")
compare_z_t_tests()
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy import stats
print(" Confidence Interval")

def confidence_interval(data, confidence=0.95): # Define the function properly
    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    ci = (mean - h, mean + h)

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
    print("Meaning: We're 95% confident the true mean lies within this range.")

# Example usage of all functions
independent_t_test()
paired_t_test()
compare_z_t_tests()
confidence_interval([8, 9, 10, 7, 9, 10, 11])
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    std_error = std_dev / np.sqrt(n)
    t_value = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_value * std_error

    print("Sample Mean:", round(mean, 2))
    print("Standard Deviation:", round(std_dev, 2))
    print("Sample Size:", n)
    print(f"Margin of Error ({int(confidence*100)}% confidence):", round(moe, 2))

# Example usage
sample_data = [50, 52, 48, 49, 51, 50, 53, 47]
margin_of_error(sample_data, confidence=0.95)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    prior_not_A = 1 - prior_A

    # Total probability of B
    prob_B = (prob_B_given_A * prior_A) + (prob_B_given_not_A * prior_not_A)

    # Bayes' Theorem: P(A|B) = (P(B|A) * P(A)) / P(B)
    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B

    print("Posterior Probability (P(A|B)):", round(posterior_A_given_B, 4))

bayes_theorem(prior_A=0.01, prob_B_given_A=0.9, prob_B_given_not_A=0.05)
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python
import pandas as pd
from scipy.stats import chi2_contingency
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female'],
    'Preference': ['Like', 'Dislike', 'Like', 'Dislike'],
    'Count': [30, 10, 20, 40]
}

df = pd.DataFrame(data)

# Create contingency table
table = pd.pivot_table(df, values='Count', index='Gender', columns='Preference')

# Chi-square test
chi2, p, dof, expected = chi2_contingency(table)

# Print results
print("Contingency Table:")
print(table)
print("\nChi-square Statistic:", round(chi2, 2))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))
print("Expected Frequencies:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))

```

```python
#16.Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f
sample1 = [10, 12, 9, 11, 13, 12, 10]
sample2 = [8, 7, 9, 6, 10, 8, 9]
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)


if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(sample1)-1, len(sample2)-1
else:
    F = var2 / var1
    dfn, dfd = len(sample2)-1, len(sample1)-1

# Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Print results
print("Variance 1:", round(var1, 2))
print("Variance 2:", round(var2, 2))
print("F-statistic:", round(F, 2))
print("p-value:", round(p_value, 4))
```

```python
#19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(groups):

    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"Normality Check for Group {i}:")
        print(f"  Shapiro-Wilk Stat = {round(stat, 3)}, p-value = {round(p_value, 4)}")
        if p_value < 0.05:
            print("  Data is NOT normally distributed (reject null hypothesis).")
        else:
            print("  Data is normally distributed (fail to reject null hypothesis).")
        print()

    # 2. Independence Check (Graphical Check using Boxplots)
    plt.figure(figsize=(8, 5))
    plt.boxplot(groups, labels=[f"Group {i}" for i in range(1, len(groups)+1)])
    plt.title('Boxplots of Groups (Independence Assumption)')
    plt.ylabel('Values')
    plt.show()

    # 3. Equal Variance Check (Levene's Test)
    stat, p_value = stats.levene(*groups)
    print(f"Equal Variance Check (Levene's Test):")
    print(f"  Levene's Test Stat = {round(stat, 3)}, p-value = {round(p_value, 4)}")
    if p_value < 0.05:
        print("  Variance is NOT equal across groups (reject null hypothesis).")
    else:
        print("  Variance is equal across groups (fail to reject null hypothesis).")

# Example usage
group1 = [22, 24, 20, 23, 25]
group2 = [30, 31, 32, 29, 30]
group3 = [40, 41, 39, 42, 43]

check_anova_assumptions([group1, group2, group3])
```

```python
#20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f
import matplotlib.pyplot as plt

# Sample data: Treatment and Gender as factors
data = {
    'Treatment': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female'],
    'Score': [23, 25, 27, 24, 22, 28, 30, 33, 31, 29]
}

# Create DataFrame
df = pd.DataFrame(data)

# Fit the Two-Way ANOVA model
model = ols('Score ~ C(Treatment) * C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print("Two-Way ANOVA Table:")
print(anova_table)

# Interpretation
if anova_table['PR(>F)']['C(Treatment)'] < 0.05:
    print("There is a significant effect of Treatment on the Score.")
else:
    print("There is no significant effect of Treatment on the Score.")

if anova_table['PR(>F)']['C(Gender)'] < 0.05:
    print("There is a significant effect of Gender on the Score.")
else:
    print("There is no significant effect of Gender on the Score.")

if anova_table['PR(>F)']['C(Treatment):C(Gender)'] < 0.05:
    print("There is a significant interaction between Treatment and Gender.")
else:
    print("There is no significant interaction between Treatment and Gender.")

# Visualizing the interaction using a boxplot
plt.figure(figsize=(8, 5))
plt.title('Interaction Effect of Treatment and Gender on Score')
plt.xlabel('Treatment and Gender')
plt.ylabel('Score')

# Boxplot for interaction effect
df['Treatment_Gender'] = df['Treatment'] + ' - ' + df['Gender']
plt.boxplot([df[df['Treatment_Gender'] == 'A - Male']['Score'],
             df[df['Treatment_Gender'] == 'A - Female']['Score'],
             df[df['Treatment_Gender'] == 'B - Male']['Score'],
             df[df['Treatment_Gender'] == 'B - Female']['Score']],
            labels=['A - Male', 'A - Female', 'B - Male', 'B - Female'])

plt.show()
```

```python
#23.Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate random data from a normal distribution
np.random.seed(42)  # For reproducibility

#  Mean = 50, Std = 10, n = 100
group1 = np.random.normal(loc=50, scale=10, size=100)

#  Mean = 52, Std = 10, n = 100
group2 = np.random.normal(loc=52, scale=10, size=100)

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

# Print results
print("t-statistic:", round(t_stat, 2))
print("p-value:", round(p_value, 4))

# Hypothesis testing interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between the means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")

# Visualizing the data with histograms and boxplots
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(group1, bins=15, alpha=0.5, label='Group 1', color='blue')
plt.hist(group2, bins=15, alpha=0.5, label='Group 2', color='orange')
plt.title('Histograms of Group 1 and Group 2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
plt.title('Boxplot Comparison of Group 1 and Group 2')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

```python
#24.D Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2

# Simulate sample data from a normal distribution
np.random.seed(42)

# Sample data: Mean = 50, Std = 10, n = 30
sample_data = np.random.normal(loc=50, scale=10, size=30)
sample_variance = np.var(sample_data, ddof=1)
sigma_0_squared = 100  # Hypothesized population variance

# Step 3: Calculate the chi-square statistic
n = len(sample_data)  # Sample size
chi_square_statistic = (n - 1) * sample_variance / sigma_0_squared

# Step 4: Degrees of freedom
df = n - 1

p_value_lower = chi2.cdf(chi_square_statistic, df)
p_value_upper = 1 - chi2.cdf(chi_square_statistic, df)
p_value = 2 * min(p_value_lower, p_value_upper)

# Step 6: Print results
print("Sample Variance:", round(sample_variance, 2))
print("Chi-square Statistic:", round(chi_square_statistic, 2))
print("p-value:", round(p_value, 4))

# Hypothesis testing interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized variance.")
```

```python
#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulate observed data: e.g., number of outcomes from a die roll (1 to 6)
np.random.seed(42)

# Simulate 600 rolls of a fair die
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=600)

# Step 1: Calculate the observed frequencies
observed_freq = np.array([np.sum(observed_data == i) for i in range(1, 7)])

# Step 2: Define the expected frequencies (for a fair die, each outcome should have equal probability)
expected_freq = np.array([600 / 6] * 6)

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(observed_freq, expected_freq)

# Step 4: Print the results
print("Observed Frequencies:", observed_freq)
print("Expected Frequencies:", expected_freq)
print("Chi-square Statistic:", round(chi2_stat, 2))
print("p-value:", round(p_value, 4))

# Step 5: Hypothesis testing interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")

# Step 6: Visualization of the observed and expected frequencies
categories = [1, 2, 3, 4, 5, 6]

plt.figure(figsize=(10, 6))
plt.bar(categories, observed_freq, alpha=0.6, label='Observed Frequencies', color='blue', width=0.4)
plt.bar(categories, expected_freq, alpha=0.6, label='Expected Frequencies', color='red', width=0.4)
plt.xlabel('Die Faces')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies for a Fair Die Roll')
plt.legend()
plt.show()
```
